## Bernoulli's Distribution

In a basket containing 100 fruits, there are 60 apples and 40 oranges. 

A fruit is picked randomly from the basket. 

1. **What is the probability that the fruit is apple?**
1. **What is the probability that it is an orange?**

````python
"""
Calculate:
p = probability that the fruit is an apple 
q = probability that the fruit is an orange
"""
apples = 60
oranges = 40
total = apples + oranges
p= apples / total
q= oranges / total

print("La probabilidad de que la fruta sea una manzana es: ", p)
print("La probabilidad de que la fruta sea una naranja es: ", q)
````

Now we take a random sample of 20 fruits from the basket. After each fruit is taken, a new fruit of the same type is replaced in the basket. Therefore, every time we are taking 1 fruit from 100 fruits. 

1. **What is the probability that the first 5 fruits are all apples?**

1. **What is the probability that the first 5 fruits are all apples and the next 15 fruits are all oranges?**

You can include the `p` and `q` probabilities you previous calculated in your solution.

````python
# What is the probability that the first 5 fruits are all apples?

apples_tries = p ** 5
oranges_tries = (p**5) * (q**15)

# Mostrar resultados con formato decimal
print(f"La probabilidad de que las primeras 5 frutas sean manzanas es: {apples_tries:.5f}")
print(f"La probabilidad de que las primeras 5 frutas sean manzanas y las siguientes 15 naranjas es: {oranges_tries:.12f}")
````

## Binomial Distribution

Similar to the above problem, this time again we randomly sample 20 fruits.

**What is the probability that this time the sample contains 5 apples and 15 oranges?**

Please note that this time the order of the fruits being picked does not matter.

````python

# your solution here
from math import comb

#Datos
n = 20  # número total de frutas
k = 5   # número de manzanas deseadas  
p = 0.6 # probabilidad de que una fruta sea manzana
q = 0.4 # probabilidad de que una fruta sea naranja

#cobinaciones
combinations = comb(n, k)

#probabilidad
probability = combinations * (p ** k) * (q ** (n - k))

print(f"La probabilidad de obtener 5 manzanas y 15 naranjas (sin importar el orden) es: {probability:.12f}")
````

In the lesson we also discussed CDF (cumulative distribution function). In the context of the above problem, **what is the probability that less than 5 fruits picked are apples?**

Write the formula and calculate the probability below.

````python
# your code here

n = 20
p=0.6
q=0.4

#Calcular P (x < 5)
prob_less_5 = sum(comb(n, i) * (p**i) * (q**(n-i)) for i in range(0, 5))

print(f"La probabilidad de que menos de 5 frutas sean manzanas es: {prob_less_5:.6f}")
````

<!-- **Plot the PDF (probability distributiom function) for the above binomial distribution.** -->

**Plot the Probability Mass Function (PMF) for the above binomial distribution.**

*Hint: The range of x in a binomial distribution is from `0` to `n`.*

````python
# your code here
# Please label the axes and give a title to the plot 

import matplotlib.pyplot as plt
from math import comb

#parametros
n = 20
p = 0.6
q = 0.4

#Calores de X: Numero de manzanas

x_values = list(range(n + 1))

#Calcular PMF: P(X = k)
pmf_values = [comb(n, k) * (p ** k) * (q ** (n - k)) for k in x_values]

#Crear el grafico

plt.bar(x_values, pmf_values, color='skyblue', alpha=0.7)
plt.xlabel("Numero de manzanas (k)")
plt.ylabel("Probabilidad P(x = k)")
plt.title("PMF de la distribucion binomial (n=20, p=0.6)")
plt.grid(alpha=0.5)
plt.show()
````

## Poisson Distribution

In this challenge you are required to use the Math library in python to solve the problems.

In the UEFA champions league tournament, the average number of goals scored in a match is 2.3. 

**What is the probability that the number of goals scored in a match is 5?**

*Hint: Use the exp() in the math library*

And to calculate the factorial, you can use the following code:

```python
import math
math.factorial(n) # n is the number you want to find the factorial of
```

````python
# your code here 

import math

lambda_goals = 2.3 # promedio de goles por partido
k = 5  

# Probabilidad de exactamente 5 goles

poisson_prob = (lambda_goals ** k * math.exp(-lambda_goals)) / math.factorial(k)
print(f"La probabilidad de que se marquen exactamente 5 goles en un partido es: {poisson_prob:.6f}")
````

**Draw a poisson probability distribution plot for the number of goals to be between 0 to 10.**

````python
# your code here
# Please label the axes and give a title to the plot 

import matplotlib.pyplot as plt

#Parametros:

x_values = list(range(0,11))

#Calcular PMF para cada valor de x

pmf_values = [(lambda_goals ** k * math.exp(-lambda_goals)) / math.factorial(k) for k in x_values]

#Crear el grafico

plt.bar( x_values, pmf_values, color='lightgreen', alpha=0.7)
plt.axvline(lambda_goals, color='red', linestyle='--', linewidth=2, label=f"λ={lambda_goals}")
plt.xlabel("Numero de goles (k)")
plt.ylabel("Probabilidad P(x = k)")
plt.title("PMF de la distribucion de Poisson (λ=2.3)")
plt.grid(alpha=0.5)
plt.xticks(x_values)
plt.show()
````
